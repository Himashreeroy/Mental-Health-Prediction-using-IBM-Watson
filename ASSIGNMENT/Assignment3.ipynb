```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Step 1: Download the dataset and load it into the tool
dataset = pd.read_csv("/kaggle/input/housing-dataset1/Housing.csv")
```

```python
# Step 2: Perform univariate analysis
# Example: Histogram of price
dataset['price'].hist()
```

```python
# Step 3: Perform bi-variate analysis
# Example: Scatter plot of price vs. area
dataset.plot.scatter('area', 'price')
```

```python
# Step 4: Perform descriptive statistics
# Example: Summary statistics
dataset.describe()
```

```python
# Step 5: Check for missing values and deal with them
dataset.isnull().sum()
```

```python
# Step 6: Find and replace outliers
# Example: Identify and replace outliers in the 'Price' column using Z-score
z_scores = np.abs((dataset['price'] - dataset['price'].mean()) / dataset['price'].std())
threshold = 3
outliers = dataset[z_scores > threshold]
dataset = dataset[z_scores <= threshold]
```

```python
# Step 7: Check for categorical columns and perform encoding
# Example: If 'Furnishing Status' is a categorical column, you can perform one-hot encoding
dataset = pd.get_dummies(dataset, columns=['furnishingstatus'])
```

```python
# Step 8: Split the data into dependent and independent variables
X = dataset.drop('price', axis=1)
y = dataset['price']
```

```python
# Step 9: Scale the independent variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 10: Split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Step 11: Build the model
model = LinearRegression()
```

```python
# Step 12: Train the model
model.fit(X_train, y_train)
```

```python
# Step 13: Test the model
y_pred = model.predict(X_test)
```

```python
# Step 14: Measure the performance using metrics
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
